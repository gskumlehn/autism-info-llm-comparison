```python
import pandas as pd

df = pd.read_csv("autism-info-llm-comparison/data.csv")
df.head()
```

```python
answers = df.iloc[:, 1:]
answers.head()
```

```python
import textstat
kincaid_grade_df = answers.map(textstat.flesch_kincaid_grade)

kincaid_grade_question_means = kincaid_grade_df.mean(axis=1)

kincaid_grade_df["Mean"] = kincaid_grade_question_means

print(kincaid_grade_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(10, 6))

sns.set_palette("crest")
sns.lineplot(data=kincaid_grade_df, linewidth=2)

plt.title("Kincaid Grade by Answer for Each Question", fontsize=16)
plt.xlabel("Question Number", fontsize=12)
plt.ylabel("Kincaid Grade", fontsize=12)
plt.xticks(ticks=range(8), labels=[str(i+1) for i in range(8)])

plt.show()
```

```python
kincaid_grade_means = kincaid_grade_df.mean()
print(kincaid_grade_means)
```

```python
plt.figure(figsize=(10, 6))
kincaid_grade_means.plot(kind='bar', color='skyblue')
plt.title('Kincaid grade average for each source')
plt.xlabel('Source')
plt.ylabel('Kincaid grade mean')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
kincaid_grade_question_means.plot(kind='bar', color='skyblue')
plt.title('Kincaid grade average for each question')
plt.xlabel('Questions')
plt.ylabel('Kincaid grade mean')
plt.xticks(ticks=range(8), labels=[str(i+1) for i in range(8)], rotation=0)
plt.tight_layout()
plt.show()
```

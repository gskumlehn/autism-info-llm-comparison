```python
import pandas as pd

df = pd.read_csv("autism-info-llm-comparison/data.csv")
df.head()
```

```python
answers = df.iloc[:, 1:]
answers.head()
```

```python
import textstat

reading_ease_df = answers.map(textstat.flesch_reading_ease)

reading_ease_question_means = reading_ease_df.mean(axis=1)

reading_ease_df["Mean"] = reading_ease_question_means

print(reading_ease_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(10, 6))

sns.set_palette("crest")
sns.lineplot(data=reading_ease_df, linewidth=2)

plt.title("Reading ease by Answer for Each Question", fontsize=16)
plt.xlabel("Question Number", fontsize=12)
plt.ylabel("Reading ease", fontsize=12)
plt.xticks(ticks=range(8), labels=[str(i+1) for i in range(8)])

plt.show()
```

```python
reading_ease_means = reading_ease_df.mean()
print(reading_ease_means)
```

```python
plt.figure(figsize=(10, 6))
reading_ease_means.plot(kind='bar', color='skyblue')
plt.title('Reading ease average for each source')
plt.xlabel('Source')
plt.ylabel('Reading ease mean')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
reading_ease_question_means.plot(kind='bar', color='skyblue')
plt.title('Reading ease average for each question')
plt.xlabel('Questions')
plt.ylabel('Reading ease mean')
plt.xticks(ticks=range(8), labels=[str(i+1) for i in range(8)], rotation=0)
plt.tight_layout()
plt.show()
```
